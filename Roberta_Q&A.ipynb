```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline

model_name = "deepset/roberta-base-squad2"
```

```python
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)
```

```python
context = "Ronaldo is widely regarded as one of the two best players of his generation, alongside Lionel Messi.[480] Winning his first Ballon d'Or in 2008 by a record-high vote count at age 23, over the next decade Ronaldo has often featured in debates concerning who is the greatest player in history.[481] Acclaimed for his prolific and consistent goal-scoring,[482] he is considered a decisive player who is also a game changer,[483] especially in important and high-pressured situations" 
nlp({
    'question': 'Who is regarded as best players?',
    'context': context
   
})
```
